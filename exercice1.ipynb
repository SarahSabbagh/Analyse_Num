```python
def p(x) :
    """"d√©finir la fonction ùëù(ùë•)=(ùë•‚àí1)6 """
    return (x-1)**6
def f(x) :
    """"d√©finir la fonction  ùëì(ùë•)=ùë•6‚àí6ùë•5+15ùë•4‚àí20ùë•3+15ùë•2‚àí6ùë•+1"""
    return x**6-6*x**5+15*x**4-20*x**3+15*x**2-6*x+1
```

```python
import numpy as np

for i in np.arange(0.998,1.002,0.0001):
    print("-f(",i,")=",f(i))
    print("-p(",i,")=",p(i))
```

On consid√®re l'√©quation :
ùëì(ùë•)=ùë•3+ùë•2‚àí3ùë•‚àí3=0
 
1) Dessiner la courbe de  ùëì  dans l'intervalle  [‚àí2;2] , puis trouver des intervalles convenables pour appliquer la m√©thode de >bisection.

2) Utiliser un programme dicho.m qui permettra de trouver les solutions de l'√©quation  ùëì(ùë•)=0  avec  ùëíùëùùë†=0.001 .

3) Modifier le programme dicho.m pour qu'il donne le nombre d'it√©ration n√©cessaire pour avoir la solution √†  ùëíùëùùë†  pr√®s.

```python
import numpy as np
import matplotlib.pyplot as plt 
x=np.linspace(-2,2,40)
g=lambda x: x**3+x**2-3*x-3
plt.plot(x,g)
plt.title('solution dans [-2,2]')
plt.show()
```

Dapr√©s le graph de  ùëì  il existe 3 solutions :
ùõº1‚àà]‚àí2;‚àí1.5[ 
ùõº2‚àà]‚àí1.5;‚àí0.5[ 
ùõº3‚àà]1.5;2[

```python
def dicho(a,b,g,tol):
    '''version it√©rative '''
    m=(a+b)/2
    err=abs(b-a)
    while err>tol :
        if g(m)==0 :
            break
        elif g(a)*g(m)<0 :
            b=m
        else: a=m
        m=(a+b)/2
        err=abs(b-a)
    return m
    
```

```python
def dichow(a,b,f,eps):
    """version tant que qui renvoila solution avec le nombre d'''it√©ration """
    m=(a+b)/2
    err=abs(b-a)
    n=0
    while err>eps :
        n+=1
        if f(m)==0 :
            break
        elif f(a)*f(m)<0 :
            b=m
        else: a=m
        m=(a+b)/2
        err=abs(b-a)
    return m,n
```
